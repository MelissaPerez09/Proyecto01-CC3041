## Análisis Empírico

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('timeResults.csv')
```

```python
data.head()
```

#### Listado de entradas de prueba usadas para medir tiempos de ejecución de la máquina.

```python
data['input'] = data['input'].apply(lambda x: str(x).count('1'))
print(data)
```

### Diagrama de dispersión que muestra los tiempos de ejecución de la máquina en función de su tamaño de entrada.

```python
plt.scatter(data['input'], data['executionTime'])

plt.xlabel('Input')
plt.ylabel('Execution Time')

plt.show()
```

### Regresión polinomial que se ajuste mejor a los datos.

```python
coeffs = np.polyfit(data['input'], data['executionTime'], 6)

poly = np.poly1d(coeffs)

x = np.linspace(min(data['input']), max(data['input']))

plt.scatter(data['input'], data['executionTime'])

plt.plot(x, poly(x), color='red')

plt.xlabel('Input')
plt.ylabel('Execution Time')

plt.show()
```
